# Mi Propuesta de proyecto: Simulación de colonia de hormigas.

### Idea
En general, la idea es generar una simulación de un grupo de hormigas que salen de su hormiguero en busca de comida. En la naturaleza se observa que las hormigas tienden a seguir un mismo camino hacia la fuente de alimento y de regreso al hormiguero. Es realmente impresionante la forma de comunicación que tiene para llegar a una organización como esa. Lo impresionante es que además, las hormigas pueden recorrer distancias larguísimas para encontrar comida, y aún asi encontrar el camino de regreso a casa. La respuesta es que, además de la memoria de algunos especímenes, la comunicación entre hormigas se da através de feromonas. Muchas (practicamente todas) especies de hormigas generan ciertos tipos de feronomas que van liberando por los lugares donde pasan (dejan un rastro). Este rastro tiene principalmente 2 funciones. Un tipo de feromona informa a otras hormigas hacia dónde se encuntra la comida (lo libera una vez que ha encontrado y va regresando), mientras que el segundo funciona como guía para regresar a la colonia (este rastro lo emite desde que sale del hormiguero). 

Una explicación más detallada para la especie **Lasius niger** se encuentra en el artículo siguente:

Czaczkes, Tomer & Castorena, Martin & Schürch, Roger & Heinze, Jürgen. (2016). Pheromone trail following in the ant Lasius niger: High accuracy and variability but no effect of task state. Physiological Entomology. 42. 10.1111/phen.12174. 

También, un estudio de diferentes feromonas en 3 especies distintas se puede consultar en este artículo:
https://www.ccubc.ca/PDF/insectsJanieChalissery.pdf

### Problema
El problema que se pretende resolver es precisamente ese, hacer que un grupo de hormigas encuentre su comida, la recoja y regrese a la colonia para llevarla con la reina. Se implementará un camino de feromonas para indicar el camino. 

Este problema tiene muchas variables involucadas, por ejemplo, la cantidad de hormigas, el tipo de feromona (de acuerdo con el segundo artículo mencionado arriba, cada sustancia segregada es más o menos intensa, por lo que la distancia que pueden recorrer las hormigas sin perder el rastro es mayor), también se pretende imponer obstáculos en el camino que tendrán que rodear, y quizá incluir depredadores.

### ¿Cómo se abordará?

Hay que tener en cuenta que son multiples los pasos intermedios que se deben hacer para llegar a la simulación.

#### Movimiento de una hormiga

Estos son problemas que se pretenden resolver simplemente para modelar el comportamiento de una hormiga en una región con otras, antes de introducir una salida común o una fuente de comida. 
Se ha observado que las hormigas, cuando no hay otras ceraca, tienden a seguir caminos en aproximadamente en linea recta para recorrer mayores distancias en busca de fuentes de alimento u otros recursos. En principio se procurará eso. Luego, si no hay necesariamente un camino de feromonas que indique que alguien ha encontrado una fuente de alimento, cuando 2 hormigas se encuentran, cambian la dirección de su avance aleatoriamente alejándose del compañero que encontraron. (https://www.youtube.com/watch?v=vG-QZOTc5_Q) (https://www.youtube.com/watch?v=gcHt5n3NGK0)

Para esto crearé una clase **Hormiga** con atributos posición (vector), vista/olfato (sección de círculo frente a su cabeza que permitirá recibir señales de su entorno), estado (que indica si lleva o no comida), entre otros que seguró tendré más claros en el momento. 
Habrá funciones auxiliares que permitirán decir si hay un encuentro con otra hormiga (además deben evitar pasar una sobre otra, aunque quiza esto no sea tan necesario), una función que indique el paso siguiente de la hormiga, también deben ser capaces de evitar obstáculos, por lo que habrá una función que les indique cómo evitar los obstáculos, entre otras funciones que permitan movilizar a una hormiga individualmente. 
Si todo sale bien y hay tiempo suficiente, además se plotearán figuras con forma de hormiga que podrían girar la cabeza y mover las patas y no solo puntos, pero esto será algo extra que sí sería bueno agregar.



El movimiento de las hormigas, antes de encontrar comida, debe ser aproximadamente en linea recta, pero no exacto, por lo que introduciré un elemento aleatorio en su movimiento para que esto ocurra más naturalmente.


#### El rastro de feromonas

Para esta parte, se pretende hacer una serie de funciones que se activen dependiendo del estado de cada hormiga. Necesitaremos funciones que permitan dejar un camino continuo de feromonas de el primer tipo y un camino continuo del segundo tipo. Además, el rastro deberá desvancerse con el tiempo. Cada hormiga debe además, una vez que consiguió la comida, regresar al hormiguero, y en caso de ser la última hormiga en tomar alimento, emitir una feromona que ayuente a las que van llegando para que ya no busque ahí. Si una hormiga recibe esta última información, la replicará.
